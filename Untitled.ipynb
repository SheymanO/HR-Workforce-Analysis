```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('Human_Resources.csv')
```

```python
df1.head()
```

```python
df1.info()
```

```python
# Convert date columns to datetime format
# Function to parse mixed date formats
def parse_dates(date):
    try:
        return pd.to_datetime(date, format='%m-%d-%y')  # Try the first format
    except ValueError:
        return pd.to_datetime(date, format='%m/%d/%Y')  # Try the second format 
    
    

# Today's date
today = pd.Timestamp('today')

    
# Apply the function to the birthdate and hire date  column
df1['birthdate'] = df1['birthdate'].apply(parse_dates)
df1['hire_date'] = df1['hire_date'].apply(parse_dates)
df1['termdate'] = pd.to_datetime(df1['termdate'].str.replace(' UTC', ''))
```

```python
# Filter out rows where birthdate is greater than today's date
df = df1[df1['birthdate'] <= today]
```

```python
df.head()
```

```python
df['termdate'].isnull()
```

```python
# Calculate age
df['age'] = (pd.Timestamp('today') - df['birthdate']).astype('<m8[Y]')
```

```python

# Calculate tenure
df['tenure'] = df.apply(
    lambda row: (row['termdate'] if pd.notnull(row['termdate']) and row['termdate'] < today else today) - row['hire_date'],
    axis=1
)
```

```python
# Convert tenure from timedelta to years
df['tenure'] = df['tenure'].astype('<m8[Y]')
```

### Turn over rate by Demographic and Job factors

```python
# Convert term_date to datetime
df['termdate'] = pd.to_datetime(df['termdate'])

# Turnover rate calculation considering future termination dates
today = pd.Timestamp('today')
df['is_terminated'] = df['termdate'].notnull() & (df['termdate'] < today)

# Convert boolean to integer
df['is_terminated'] = df['is_terminated'].astype(int)
df
```

```python
# Turnover rate by gender
turnover_by_gender = df.groupby('gender')['is_terminated'].mean()
print(turnover_by_gender)


```

```python
# Turnover rate by race
turnover_by_race = df.groupby('race')['is_terminated'].mean()
print(turnover_by_race)

```

```python
# Turnover rate by department
turnover_by_department = df.groupby('department')['is_terminated'].mean()
turnover_by_department.sort_values(ascending = False)

```

```python
# Turnover rate by job title
turnover_by_jobtitle = df.groupby('jobtitle')['is_terminated'].mean()
turnover_by_jobtitle.sort_values(ascending = False)

```

```python
# Turnover rate by location
turnover_by_location = df.groupby('location')['is_terminated'].mean()
turnover_by_location.sort_values(ascending = False)
```
